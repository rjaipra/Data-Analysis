```python
#Complaint type Breakdown with bar plot to figure out majority of complaint types and top 10 complaints
ds['Complaint Type'].value_counts().plot(kind='bar',title='count vs. complaint types',alpha=1.0)#plot(kind='barh',alpha=0.6,figsize=(15,30))
plt.show()
```

```python
#top 10 complaints
ds['Complaint Type'].value_counts().nlargest(10).plot(kind='bar',title='count vs. complaint types',alpha=1.0)
```

```python
#Status Vs Count plot
ds['Status'].value_counts().plot(kind='bar',title='Status vs. Count',alpha=1.0)
```

```python
#top 10 complaining cities
ds['City'].value_counts().nlargest(10).plot(kind='bar',title='count vs. city',alpha=1.0)
```
